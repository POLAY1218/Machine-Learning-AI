<font size="3"><b>Bernoulli trial Simulation(random coin flip experiment where we can define the bias of our coin )</b></font><br>
<font size="3">1. After running the Bernoulli simulation  for nth times and verify if the end result from our simulation conforms w our defined bias</font><br>
<font size="3">2. We will run the Bernoulli experiment to generate data for the rv as well as find the relationship between the mean of those data and the probability of success as the step towards simulating law of large numbers(llm) for Bernoulli distribution.</font><br>
<font size="3">3. And finally build a simulation of Law of large numbers which would show that increasing the sample size results sample mean to converge to the expected value.</font><br>


```python
import numpy as np
```

```python
"""Define a bernoulli function to simulate a single coin flip where we can set the bias of the coin. The output of the function tell and  us result of the single coin flip ie head(1) or tail(0)."""

def BernoulliTrial(p=0.5):
    X = int(np.random.rand()<=p)
    return X

print(BernoulliTrial(0.7))
```

```python
"""Next we would do a simulation where we would flip this coin say n times and find 
if the the end result(fracion of success and fraction of failure) is close to the bias of coin; ps: the higher the n the closer 
the simulation should confirm the bias of our coin """

countOnes = 0
countZeros = 0
p = 0.7
n = 100000
for i in range(n):
    x = BernoulliTrial(p)
    if x == 1:
        countOnes += 1
    else:
        countZeros += 1
print("Success Fraction is : ",countOnes/n)
print("Failure Fraction is : ",countZeros/n)
```

<font size="3">So abve its clear that the end result (fraction of success and fraction of failure) of the simulation conforms w/ our set bias.</font></br>
<font size="3"><b>Next we will run the Bernoulli experiment to generate data for the rv and find the relationship between the mean of those data and the probability of success as the step towards simulating law of large numbers(llm) for Bernoulli distribution..</b></font>

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)

p = 0.71 # probability of success
N = 1000000 # numbers of Bernoulli trials
X= (np.random.rand(N)<=p) 
print(X)
```

```python
X.mean() 
```

<font size="3">So above, the mean we get  is very close to the probability of success we assigned; the result is showing .709696 vs .71   ( this shows 73% of the bernoulli trials would give us heads out of 1,000000 trials and the rest  apprx .27% would be tail.)</font></br>
<font size="3">This means is that for Bernoulli random variable the mean and the parameter p(probability of success) of its pmf are the same. We will use this fact to simulate Law of large numbers(LLM) next.</font></br>
<font size="3"><b>Next we will build a simulation to help us realize the fundamental law large numbers in the context of Bernoulli rv which says increasing the sample size will result sample mean to converge to population mean ie expected value,E(X)</b></font>

```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.812
print(f"E(X)=p = {p}")
n = 10000

sMeanX = []
for N in range(1, n, 100):
    B = [BernoulliTrial(p) for _ in range(N)]
    sMeanX.append(np.mean(B))

y = np.array(sMeanX)
x = np.arange(1, n, 100)

plt.figure(figsize=(10,5))
plt.plot(x, y, linestyle='--', label='Simulated Mean')
plt.axhline(y=p, color='r', linestyle='-', label='Population Mean or Expected Value ie E(x)= p')
plt.title('Simulation of Law of Large Numbers for Bernoulli random variables')
plt.xlabel('Sample size')
plt.ylabel('Sample Mean')
plt.legend()
plt.show()
```

<font size="4">Above we can clearly see as the sample size increases the variations in mean values keep getting smaller and eventually converges to its population mean or expected value</font>


<font size="3"><b>Geometric random variable simulation of the coin flip experiment(flipping a coin until it returns a 'head' )</b></font><br>
<font size="3">1.First we would  simulate Geometric random variable using the Bernoulli function we defined earlier</font><br>
<font size="3">2. We will run the Geometric experiment to generate data for the rv as well as find the relationship between the mean of those data and the probability of success as the step towards simulating law of large numbers(llm) for Geometric distribution.</font><br>
<font size="3">3. And finally build a simulation of Law of large numbers which would show that increasing the sample size results sample mean to converge to the expected value.</font><br>


```python
"""We would define a function that would run the coin flip trial simulation until it gets the 'heads'. Output would tell us the total number of flips which is the return value of the Geometric random variable"""
def GT(p):
    
    Y = 1 
    
    while True:
        
        if BernoulliTrial(p) == 1:
            
            return Y
        else:
            Y+=1 
            
    
p = 0.5
print(GT(p))
```

```python
"""Running the geometric simulation trials for nth times(coin flip experiment w n number of trials where each trial ends when it returns the'heads') .After that each return value of the trial would be stored  in an array .
Fianlly we would plot a graph on the output array """

n = 1000000 
p = 0.6

G = np.zeros(n) 


for i in range(n):
    G[i] = GT(p)

G
```

```python
"""Plotting graph on the result of the Geometric Simulation experiment(ie output array above)"""

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.distplot(G)
```

<font size="4">the plot above is showing us the frequency of individual trials to return a 'head'.Next we will increase the total number of trials(n) and decrease the probability for better visualization and demonstration 

```python
n = 100000
p = 0.02
G = np.zeros(n)
for i in range(n):
    G[i] = GT(p)
    
    
sns.distplot(G)
```

<font size="4">Above graph is  a clear demonstration of Geometric random variable and its decay

```python
"""We could  do a  histogram w proper density for the geometric simulation experiment"""
n = 100000
p = 0.6
G = np.zeros(n)
for i in range(n):
    G[i] = GT(p)
    

plt.hist(G,density=True)
```

<font size="4">so now that the density is properly scaled we can see that almost 70 % of total mass is being accumulated between 1 and 2 flips since our probability (p) is pretty high; lets set the probability (p) to a really low  value and run the simulation and then  plot the simulation again

```python
n = 100000
p = 0.02
G = np.zeros(n)
for i in range(n):
    G[i] = GT(p)
    

plt.hist(G,density=True)
```

<font size="4">Now above  we see that only about 15% of the mass is being accumulated between 0 and 100 flips;We can measure the approximate probability distribution of a  geometric random variable using histogram this way

```python
"""Earlier we generated data by running Geometric rv simulation, we will calculate the mean  to find the relationship between the mean and its assigned probability(p=.6)"""
sMeanX=G.mean()
sMeanX
```

<font size="4">So above we can see that the mean value of the dataset from Geometric experiment is actually the same as the inverse value of its probability of success(since 1/p= 1/.6=1.666) . We will use this inverse relationship between the mean the probability of success as the step towards simulating law of large numbers(llm) for Geometric distribution.</font><br>
 
<font size="4">Next we will build a simulation to help us realize the fundamental law large numbers in the context of Geometric rv which says increasing the sample size will result sample mean to converge to population mean ie expected value,E(X)

```python
import numpy as np
import matplotlib.pyplot as plt

def BernoulliTrial(p=0.5):
    X = int(np.random.rand()<=p)
    return X

def GT(p):
    Y = 1 
    while True:
        if BernoulliTrial(p) == 1:
            return Y
        else:
            Y += 1 

p = 0.73
print(f"E(X)=1/p = {1/p}")
n = 10000

sMeanX = []
for N in range(1, n, 100):
    G = [GT(p) for _ in range(N)]
    sMeanX.append(np.mean(G))

y = np.array(sMeanX)
x = np.arange(1, n, 100)

plt.figure(figsize=(10,5))
plt.plot(x, y, linestyle='--', label='Simulated Mean')
plt.axhline(y=1/p, color='r', linestyle='-', label='Population Mean or Expected value, E(X)=1/p')
plt.title('Simulation of Law of Large Numbers for geometric rv')
plt.xlabel('Sample size')
plt.ylabel('Sample Mean')
plt.legend()
plt.show()
```

<font size="4">Above we can clearly see as the sample size increases the variations in mean values keep getting smaller and eventually converges to its population mean or expected value</b></font>

<font size="4"><b>Define a Binomial random variable to simulate a coin flip experiment (flip a fair or biased coin nth times and find the total number of heads, then repeat this experiment for Nth times and find the distribution)

```python
"""Define a function where we can set the parameters(n) for  the total number of coin flips and the probability threshold(coin bias). The ouput of the function would tell us the total number of coin flips(n) and the 
total number of 'heads' ."""
np.random.seed(0)
def BT(n,p):
    X = (np.random.rand(n)<=p).sum() 
    return X

print(BT(100,0.2))
```

```python
"""Now we would run the binomial  simulation Nth times and store the values of each experiment in an array. Next we would plot a graph on the simulation array output"""

N = 10000 
n = 1000
p = 0.5
B = np.zeros(N)
for i in range(N):
    B[i] = BT(n,p)
B
```

```python
"""We would plot a  histogram and KDEdensity plot on the histogram(to show continous line) on our simulation array output"""
plt.hist(B,density=True,bins=20)
sns.kdeplot(B,shade=True)
```

```python
"""Earlier we generated data by running binomial rv simulation, we will calculate the mean to find the relationship between the mean and its assigned probability(p=.5)"""
sMeanX=B.mean()
sMeanX
```

<font size="4">So above we can see that the mean value of the dataset from Binomial experiment is actually the product number of trials(n) and  the probability(p) of success(since n.p= 1000*.5=500 vs sMeanX=499.8212) . We will use this relationship between the mean and 'number of trial*probability of success ' as the step towards simulating law of large numbers(llm) for binomial distribution.
 
<font size="4">Next we will build a simulation to help us realize the fundamental law large numbers in the context of Binomial rv which says increasing the sample size will result sample mean to converge to population mean ie expected value,E(X)

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10000 # Total number of trials
n = 1000 # Parameter for the binomial distribution
p = 0.6 
B = np.zeros(N) 
sMeanX = [] 
print(f"E(X)=n*p = {n*p}")
for i in range(N):
    B[i] = BT(n,p)
    sMeanX.append(B[:i+1].mean())

y = np.array(sMeanX)
x = np.arange(1, N+1)

plt.figure(figsize=(10,5))
plt.plot(x, y, linestyle='--', label='Simulated Mean')
plt.axhline(y=n*p, color='r', linestyle='-', label='Population Mean or Expected value, E(X)=np')
plt.title('Simulation of Law of Large Numbers for binomial rv')
plt.xlabel('Sample size')
plt.ylabel('Sample Mean')
plt.legend()
plt.show()
```

<font size="4">Above we can clearly see as the sample size increases the variations in mean values keep getting smaller and eventually converges to its population mean or expected value</b></font>

<font size="4"><b>Simulation of exponential random variable</b></font><br>

<font size="4">Exponential random variable is a continuous probability distribution used to model the time between events assuming that these events occur independently at a constant average rate.It  is often used to model the time between occurrences of events in a Poisson  process(we will see Poisson rv next).</font><br>
<font size="4">Example:: Suppose buses arrive at a bus stop at an average rate of one every 15 minutes, and the times between buses are independent, then the waiting time until the next bus can be modeled as an exponential random variable.


```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10,1000)
lambdas = np.linspace(1,2,4)
for i in lambdas:
    fx = i*np.exp(-i*x)
    plt.plot(x, fx, label=str(i))
    plt.fill_between(x, fx, alpha=0.1) 

plt.legend()
plt.show()
```

<font size="4">So abv we see smaller lambda giving lower peak and slower decay and larger lambda giving higher peak and faster decay. Notice 3 of the pdf has a value higher than 1, but thats not an issue as long as the area under each of those curves remains 1 which obviously would to hold the normalization axiom of probability . 

<font size="4"><b>Simulation of Poisson random variable</b></font><br>
<font size="4">Poisson random variable  represents the number of events in a fixed interval of time or space in a Poisson process. A Poisson process is a type of mathematical model that describes events occurring randomly in time and/or space where events occur discretely, independently of each other and at a  constant rate  . Exponential rv is used to model the time between events in the poisson process.


```python
import numpy as np
import matplotlib.pyplot as plt


lmbda = 5 # average rate of events
time_limit = 10

poisson_rv = []

# Generate Poisson random variables
for _ in range(10000):
    time = 0
    count = 0
    while time < time_limit:
        # Generate an exponential random variable (waiting time until next event)
        waiting_time = np.random.exponential(scale=1/lmbda)
        time += waiting_time
        if time < time_limit:
            count += 1
    poisson_rv.append(count)

plt.hist(poisson_rv, bins=np.arange(min(poisson_rv), max(poisson_rv)+1),
         align='left', color='lightblue', edgecolor='black', density=True)

plt.title("Poisson Distribution with  = {}".format(lmbda))
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.show()
```

<font size="4">Due to  CLT  as the average rate increases  the shape of poisson distribution looks increasingly gaussian. For better illustration of CLT lets plot it using different values of rate of events

```python
import numpy as np
import matplotlib.pyplot as plt

for lmbda in [5, 30, 100]:
    poisson_rv = np.random.poisson(lam=lmbda, size=10000)

    plt.hist(poisson_rv, bins=np.arange(min(poisson_rv), max(poisson_rv)+1),
             align='left', color='skyblue', edgecolor='black', density=True, alpha=0.5, label=f' = {lmbda}')

plt.title("Poisson Distribution")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.legend()
plt.show()
```
