##  Importing and Saving the Movies Dataset (Best Practice)

```python
import pandas as pd
import requests
import json
pd.options.display.max_columns = 30
```

```python
movie_id = [0, 299534, 19995, 140607, 299536, 597, 135397,
            420818, 24428, 168259, 99861, 284054, 12445,
            181808, 330457, 351286, 109445, 321612, 260513]
```

```python
basic_url = 'https://api.themoviedb.org/3/movie/{}?{}' 
```

```python
json_list = []
for movie in movie_id:
    url = basic_url.format(movie, api_key)
    r = requests.get(url)
    if r.status_code != 200:
        continue
    else:
        data = r.json()
        json_list.append(data) 
df = pd.DataFrame(json_list)
```

```python
requests.get(basic_url.format(0, api_key)).status_code
```

```python
df
```

```python
df = df.loc[:, ["title", "id", "revenue", "genres", "belongs_to_collection", "runtime"]].sort_values(by = "revenue",
                                                                                                ascending = False)
```

```python
df
```

```python
df.to_json("movies.json", orient = "records")
```

## Normalizing Unicode Strings and Getting the Ticker Symbols

```python
import unicodedata
```

```python
const.Symbol
```

```python
const.Symbol.to_list()
```

```python
const.info()
```

```python
const.Symbol.apply(lambda x: unicodedata.normalize("NFKD", x))
```

```python
const.Symbol = const.Symbol.apply(lambda x: unicodedata.normalize("NFKD", x))
```

```python
const.Symbol[0]
```

```python
const["Ticker"] = const.Symbol.str.split(": ").apply(lambda x: x[-1])
```
